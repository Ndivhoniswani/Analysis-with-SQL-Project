```python
Question 1
Who won the Oscar for “Actor in a Leading Role” in 2015?

(Hint: The winner is indicated as '1.0'.)
```

```python
%%sql

SELECT name
FROM oscars
WHERE winner = 1.0
AND award = "Actor in a Leading Role"
AND year = 2015
LIMIT 5;
```

```python
Question 5
How many movies are there that are both in the "Thriller" genre and contain the word “love” anywhere in the keywords?
```

```python
%%sql

SELECT COUNT(movies.title) AS num_movies

FROM movies
JOIN genremap ON movies.movie_id = genremap.movie_id
JOIN genres  ON genremap.genre_id = genres.genre_id

JOIN keywordmap ON movies.movie_id = keywordmap.movie_id
JOIN keywords ON keywordmap.keyword_id = keywords.keyword_id

WHERE genres.genre_name = "Thriller"
AND keywords.keyword_name = 'love';
```

```python
Question 6
How many movies are there that were released between 1 August 2006 ('2006-08-01') and 1 October 2009 ('2009-10-01') that have a popularity score of more than 40 and a budget of less than 50 000 000?
```

```python
%%sql

SELECT COUNT(title) AS num_movies

FROM movies
WHERE release_date between '2006-08-01' AND '2009-10-01'
AND popularity > 40
AND budget < 50000000;
```

```python
Question 7
How many unique characters has "Vin Diesel" played so far in the database?
```

```python
%%sql

SELECT COUNT(DISTINCT casts.characters) AS unique_characters

FROM casts
JOIN actors ON casts.actor_id = actors.actor_id
WHERE actors.actor_name = 'Vin Diesel';
```

```python
Question 8
What are the genres of the movie “The Royal Tenenbaums”?
```

```python
Question 9
What are the three production companies that have the highest movie popularity score on average, as recorded within the database?
```

```python
%%sql

SELECT productioncompanies.production_company_name,
       AVG(movies.popularity) AS Average_popularity


FROM movies
JOIN productioncompanymap ON movies.movie_id = productioncompanymap.movie_id
JOIN productioncompanies ON productioncompanymap.production_company_id = productioncompanies.production_company_id


GROUP BY productioncompanies.production_company_name
ORDER BY AVG(movies.popularity) DESC
LIMIT 3;
```

```python
Question 10
How many female actors (i.e. gender = 1) have a name that starts with the letter "N"?
```

```python
%%sql

SELECT SUM(gender) AS num_of_female_actors
FROM actors
WHERE gender = 1 AND actor_name LIKE 'N%';
```

```python
Question 11
Which genre has, on average, the lowest movie popularity score?
```

```python
%%sql

SELECT genres.genre_name,
       AVG(movies.popularity) AS Average_popularity

FROM movies
JOIN genremap ON movies.movie_id = genremap.movie_id
JOIN genres  ON genremap.genre_id = genres.genre_id

GROUP BY genres.genre_name
ORDER BY AVG(movies.popularity) ASC
LIMIT 5;
```

```python
Question 12
Which award category has the highest number of actor nominations (actors can be male or female)? (Hint: Oscars.name contains both actors' names and film names.)
```

```python
%%sql

SELECT award, COUNT(name) AS num_nominations
FROM Oscars

JOIN actors
ON Oscars.Name = actors.actor_name

GROUP BY award
ORDER BY COUNT(name) DESC
LIMIT 2;
```

```python
Question 14
DStv will be having a special week dedicated to the actor Alan Rickman. Which of the following queries would create a new view that shows the titles, release dates, taglines, and overviews of all movies that Alan Rickman has played in?
```

```python
%%sql

CREATE VIEW Alan_Rickman_Movies AS
SELECT title, release_date, tagline, overview FROM Movies
LEFT JOIN Casts ON Casts.movie_id = Movies.movie_id Left JOIN Actors ON Casts.actor_id = Actors.actor_id WHERE Actors.actor_name = 'Alan Rickman'
```
